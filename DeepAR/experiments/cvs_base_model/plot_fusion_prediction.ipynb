```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime

# 读取CSV文件
df = pd.read_csv('fusion_predictions.csv')

# 将Date列转换为datetime类型
df['Date'] = pd.to_datetime(df['Date'])

# 创建图表
plt.figure(figsize=(12, 6))

# 绘制actual和prediction的折线图
plt.plot(df['Date'], df['actual'], '-', label='Actual', color='blue', linewidth=1.5)
plt.plot(df['Date'], df['prediction'], '--', label='Prediction', color='red', linewidth=1.5)

# 设置x轴格式，每个月显示一个刻度
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))

# 旋转日期标签，使其更易读
plt.xticks(rotation=45)

# 添加标题和标签
plt.title('Actual vs Fusion Prediction Stock Returns')
plt.xlabel('Date')
plt.ylabel('Daily Return')
plt.legend()
plt.grid(True, alpha=0.3)

# 调整布局
plt.tight_layout()

# 保存图表
plt.savefig('fusion_prediction_plot.png', dpi=300)

# 显示图表
plt.show()
```
