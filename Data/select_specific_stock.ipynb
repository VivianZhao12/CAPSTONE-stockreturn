```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/Users/amily/Desktop/stock.csv")
df['Date'] = pd.to_datetime(df['Date'])

# Filter for Amazon (AMZN) stocks within the date range
filtered_df = df[
    (df['Ticker'] == 'AMZN') & 
    (df['Date'] >= '2021-12-31') & 
    (df['Date'] <= '2023-01-01')
]

filtered_df['Daily_Return'] = (filtered_df['Close'] - filtered_df['Close'].shift(1)) / filtered_df['Close'].shift(1) 
#filtered_df[["Date","Ticker","Industry"]]
filtered_df = filtered_df[
    (df['Ticker'] == 'AMZN') & 
    (df['Date'] >= '2022-01-01') & 
    (df['Date'] <= '2023-01-01')
]
filtered_df.to_csv("/Users/amily/Desktop/amazon_stock.csv")
filtered_df
```
